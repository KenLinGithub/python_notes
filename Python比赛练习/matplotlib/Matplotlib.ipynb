# Matplotli教程

````python
import matplotlib.pyplot as plt
````

## 线型图
`plot(x, y, [fmt], linewidth)`

````python
x = range(1, 6)
y = [1, 9, 3, 7, 5]

plt.plot(x, y)

plt.show()
````

## 线条颜色、标记、线形样式
### 颜色字符
* 蓝色`b`
* 洋红色`m`
* 绿色`g`
* 黄色`y`
* 红色`r`
* 黑色`k`
* 白色`w`
* 青绿色`c`
* HEX颜色`#42A5F5`

多条曲线不指定颜色时，会自动选择不同颜色。


### 线型字符
* 实线`-`
* 虚线`.`
* 破折线`--`
* 点划线`-.`

### 标记字符
* 点标记`.`
* 像素标记(极小点)`,`
* 实心圈标记`o`
* 倒三角标记`v`
* 上三角标记`^`
* 右三角标记`<`
* 左三角标记`>`

````python
x = range(1, 6)
y = [1, 9, 3, 7, 5]

plt.plot(x, y, 'g-.o', linewidth=4)

plt.show()
````

## 柱状图
* 柱状图`plt.bar()`
* 横向柱状图`plt.barh()`

````python
x = range(1, 6)
y = [1, 9, 3, 7, 5]

plt.barh(x, y)

plt.show()
````

## 饼图
```python
plt.pie(values,
        explode,
        labels,
        colors,
        autopct,
        startangle,
        counterclock,
       )
```

### values 扇形面积

### explode 突出显示
值越大，距离中心越远，默认为 0

### labels 扇形标签
默认为None

### autopct 扇形百分比显示格式
* 整数`%d`
* 整数百分比`%d%%`
* 一位小数`%0.1f`
* 一位小数百分比`%0.1f%%`

### stratangle 起始绘图角度
默认为0
* x 正半轴`0`
* y 正半轴`90`
* x 负半轴`180`
* y 负半轴`270`

### counterclock 指针方向
* 逆时针`True`
* 顺时针`False`
默认为 True

````python
labels = ['A', 'B', 'C', 'D', 'E']
values = [1, 9, 3, 7, 5]

plt.pie(values, labels = labels)

plt.axis('equal')
plt.show()
````

````python
labels = ['A', 'B', 'C', 'D', 'E']
values = [1, 9, 3, 7, 5]

plt.pie(values,
        explode = (0, 0.1, 0, 0, 0),
        labels = labels,
        autopct = "%0.2f%%",
        startangle = 90,
        counterclock = False,
       )

plt.axis('equal') #等比例

plt.show()
````

## 其他组件

### 标题
`plt.title('Test Plot')`

### x 轴标签
`plt.xlabel('X')`

### y 轴标签
`plt.ylabel('Y')`

### 图例
`plt.legend(['Y', 'Z'])`

### x 轴范围
`plt.xlim(0, 10)`

### y 轴范围
`plt.ylim(0, 20)`

````python
x = range(1, 6)
y = [1, 9, 3, 7, 5]
z = [5, 7, 3, 9, 1]

plt.plot(x, y)
plt.plot(x, z)
plt.title('Test Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['Y', 'Z'])

plt.xlim(0, 8)
plt.ylim(0, 12)
plt.show()
````

## 内置样式

````python
plt.style.available
````

````python
x = range(1, 6)
y = [1, 9, 3, 7, 5]
z = [5, 7, 3, 9, 1]

for style in plt.style.available:
    with plt.style.context(style):
        plt.plot(x, y)
        plt.plot(x, z)
        plt.title(style)
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.legend(['Y', 'Z'])

        plt.xlim(0, 8)
        plt.ylim(0, 12)
        plt.show()
````

````python
x = range(1, 6)
y = [1, 9, 3, 7, 5]



for style in plt.style.available:
    with plt.style.context(style):
        plt.bar(x, y)
        plt.title(style)
        plt.ylabel('Y')
        plt.xlim(0, 8)
        plt.ylim(0, 12)
        plt.show()
````
